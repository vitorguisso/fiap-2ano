Nesta etapa, foi realizado o carregamento da base de dados contendo frases com sintomas e seus respectivos níveis de risco. O arquivo no formato CSV foi importado utilizando a biblioteca pandas, permitindo a manipulação estruturada dos dados. Em seguida, foram exibidas as primeiras linhas da base para validação do conteúdo, bem como a distribuição das classes por meio da contagem de ocorrências. Observou-se que a base está balanceada, com 12 exemplos de “alto risco” e 12 de “baixo risco”, o que contribui para um treinamento mais adequado do modelo de Machine Learning

```python
import pandas as pd
from google.colab import files

print("Faça o upload do arquivo risco.csv")
uploaded = files.upload()

df = pd.read_csv("risco.csv")

print("\nBase carregada com sucesso.")
print("\nPrimeiras linhas da base:")
print(df.head())

print("\nQuantidade de exemplos por classe:")
print(df["risco"].value_counts())
```

Nesta etapa, os dados foram preparados para o treinamento do modelo de Machine Learning. As frases previamente convertidas em vetores numéricos por meio da técnica TF-IDF foram divididas em conjuntos de treino e teste utilizando a função train_test_split da biblioteca Scikit-learn. Foi utilizada uma proporção de 70% dos dados para treino e 30% para teste. Além disso, o parâmetro stratify foi aplicado para garantir que a distribuição das classes (alto risco e baixo risco) fosse mantida em ambos os conjuntos. O parâmetro random_state foi definido para assegurar a reprodutibilidade dos resultados.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

X = df["frase"]
y = df["risco"]

vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

print("Dados preparados com sucesso.")
print("Quantidade para treino:", X_train.shape[0])
print("Quantidade para teste:", X_test.shape[0])
```

Nesta etapa, foi utilizado o algoritmo de Regressão Logística para realizar a classificação das frases em níveis de risco. O modelo foi treinado a partir dos dados previamente preparados, onde as frases foram convertidas em representações numéricas por meio da técnica TF-IDF. A função fit foi utilizada para ajustar o modelo aos dados de treino, permitindo que o algoritmo aprendesse padrões linguísticos associados a cada classe. A Regressão Logística foi escolhida por sua eficiência e bom desempenho em tarefas de classificação de texto.

```python
from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression()
modelo.fit(X_train, y_train)

print("Modelo treinado com sucesso.")
```

O modelo foi avaliado utilizando métricas clássicas de classificação, incluindo acurácia, relatório de classificação e matriz de confusão. A acurácia obtida foi de 87,5%, indicando um bom desempenho na tarefa de classificação. O relatório demonstrou que o modelo apresentou recall de 1,00 para a classe de alto risco, ou seja, foi capaz de identificar corretamente todos os casos considerados críticos. A matriz de confusão revelou apenas um erro de classificação, no qual um caso de baixo risco foi classificado como alto risco. Esse comportamento pode ser considerado aceitável em contextos de triagem médica, onde é preferível superestimar o risco do que deixar de identificar um caso potencialmente grave.

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = modelo.predict(X_test)

print("ACURÁCIA DO MODELO:")
print(accuracy_score(y_test, y_pred))

print("\nRELATÓRIO DE CLASSIFICAÇÃO:")
print(classification_report(y_test, y_pred))

print("\nMATRIZ DE CONFUSÃO:")
print(confusion_matrix(y_test, y_pred))
```

Por fim, foi implementado um sistema de teste manual, permitindo a inserção de novas frases pelo usuário. Essas frases são transformadas em vetores numéricos utilizando TF-IDF e classificadas pelo modelo treinado. Os resultados demonstraram que o modelo é capaz de generalizar padrões aprendidos, classificando corretamente diferentes tipos de sintomas como “alto risco” ou “baixo risco”. Essa etapa simula a aplicação prática do sistema em um cenário real de triagem automatizada.

```python
while True:
    frase = input("Digite uma frase com sintomas ou 'sair': ").strip()

    if frase.lower() == "sair":
        print("Encerrando testes.")
        break

    frase_tfidf = vectorizer.transform([frase])
    previsao = modelo.predict(frase_tfidf)[0]

    print(f"Classificação prevista: {previsao}\n")
```

Durante os testes manuais, foi possível observar limitações do modelo ao lidar com frases contendo sintomas que não estavam presentes na base de treinamento. Por exemplo, ao inserir a frase “febre e sede”, o modelo classificou como “alto risco”, mesmo não sendo um padrão aprendido diretamente. Isso ocorre porque o modelo tenta generalizar com base em padrões existentes, podendo gerar classificações imprecisas em casos fora do domínio da base. Esse comportamento evidencia a importância de uma base de dados mais abrangente para melhorar a capacidade de generalização do modelo.
