3) Simulação de pacientes

```python
import pandas as pd
import numpy as np

# Ler base original
df = pd.read_csv("heart_disease_uci.csv")

# Quantos pacientes simulados você quer
n_simulados = 1000

# Separar colunas numéricas e categóricas
colunas_numericas = ["age", "trestbps", "chol", "thalch", "oldpeak", "ca"]
colunas_categoricas = ["sex", "dataset", "cp", "fbs", "restecg", "exang", "slope", "thal", "num"]

# Criar DataFrame vazio
simulados = pd.DataFrame()

# Gerar colunas numéricas com base na média e desvio padrão da base real
for col in colunas_numericas:
    media = df[col].mean()
    desvio = df[col].std()
    simulados[col] = np.random.normal(media, desvio, n_simulados)

# Ajustes para evitar valores estranhos
simulados["age"] = simulados["age"].round().clip(28, 77)
simulados["trestbps"] = simulados["trestbps"].round().clip(80, 220)
simulados["chol"] = simulados["chol"].round().clip(80, 700)
simulados["thalch"] = simulados["thalch"].round().clip(60, 220)
simulados["oldpeak"] = simulados["oldpeak"].round(1).clip(0, 7)
simulados["ca"] = simulados["ca"].round().clip(0, 4)

# Gerar colunas categóricas respeitando a proporção observada no dataset
for col in colunas_categoricas:
    distribuicao = df[col].dropna().value_counts(normalize=True)
    simulados[col] = np.random.choice(
        distribuicao.index,
        size=n_simulados,
        p=distribuicao.values
    )

# Criar id novo
simulados.insert(0, "id", range(10001, 10001 + n_simulados))

# Reordenar colunas
simulados = simulados[
    ["id", "age", "sex", "dataset", "cp", "trestbps", "chol", "fbs",
     "restecg", "thalch", "exang", "oldpeak", "slope", "ca", "thal", "num"]
]

# Visualizar
print(simulados.head())

# Salvar
simulados.to_csv("heart_disease_simulado.csv", index=False)

print("Arquivo gerado com sucesso: heart_disease_simulado.csv")
```
